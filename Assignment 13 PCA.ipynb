# Hierarchiel Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
import scipy.cluster.hierarchy as sch
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler
scaled=StandardScaler()
```

```python
df1=scaled.fit_transform(df1)
df1
```

```python
dfh=pd.DataFrame(df1,columns=['pca1','pca2','pca3'])
dfh
```

```python
dendrogram=sch.dendrogram(sch.linkage(dfh,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y_hc=hc.fit_predict(dfh)
Clusters=pd.DataFrame(y_hc,columns=['clusters'])
```

```python
Clusters.value_counts()
```

```python
dfh['Hierarchiel clusters']=y_hc
```

```python
dfh
```

# KMeans

```python
dfk=pd.DataFrame(df1,columns=['pca1','pca2','pca3'])
dfk
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=5)
    kmeans.fit(dfk)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.show()
```

```python
clusters=KMeans(3,random_state=5)
```

```python
clusters.fit(dfk)
```

```python
clusters.labels_
```

```python
dfk['KMeans clusters']=clusters.labels_
dfk
```

```python
dfk['KMeans clusters'].value_counts()
```
